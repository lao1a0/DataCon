```python
import pandas as pd

df = pd.read_excel('DNS.xlsx')
df
```

```python
info_split = df['Info'].str.split(' ', expand=True)
info_split[0] = info_split[0] + ' ' + info_split[1]
info_split = info_split.drop(columns=[1])

# 为新的列添加列名
info_split.columns = ['query_id', 'record_type', 'domain_name', 'response_type', 'ip_address_1', 'ip_address_2', 'ip_address_3']
info_split
```

```python
df = pd.concat([df, info_split], axis=1)
df
```

```python
df.drop(columns=['query_id'], inplace=True)
```

```python
df
```

```python
merged_df = pd.DataFrame()

# 遍历数据框，合并满足条件的行
i = 0
while i < len(df) - 1:
    if df.loc[i, 'record_type'] == df.loc[i + 1, 'domain_name']:
        # 合并第i行和第i+1行
        merged_row = pd.concat([df.iloc[i], df.iloc[i + 1]], axis=0)
        merged_df = merged_df.append(merged_row, ignore_index=True)
        i += 2  # 跳过下一行
    else:
        # 如果不满足条件，直接添加第i行
        merged_df = merged_df.append(df.iloc[i], ignore_index=True)
        i += 1

# 如果最后一行没有被合并，直接添加
if i == len(df) - 1:
    merged_df = merged_df.append(df.iloc[i], ignore_index=True)
```

```python
merged_df.columns=['No', 'Protocol', 'Length', 'Info', 'record_type', 'domain_name',
       'response_type', 'ip_address_1', 'ip_address_2', 'ip_address_3', 
       'No2','Protocol2', 'Length2', 'Info2', 'record_type2', 'domain_name2',
       'response_type2', 'ip_address_12', 'ip_address_22', 'ip_address_32']
```
